### Matplotlib Below

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,9))
x = np.arange(-10,11)
y = x**2
plt.plot(x,y)
```

```python
plt.figure(figsize=(16,9))
plt.plot(x,y)
plt.grid()
plt.title("График $y=x^2$")
plt.xlabel("x")
plt.ylabel("y")
```
