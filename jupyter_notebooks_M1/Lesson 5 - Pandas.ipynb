```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
df.groupby('city').count().sort_values(by='id', ascending=False).head()
```

```python
result = df[df['name'] == 'Taco Bell']['city'].value_counts().head()
result
```

```python
df['dateAdded'] = pd.to_datetime(df['dateAdded'])
df[df['dateAdded'].dt.month == 11].head().index
```

```python
result = df.groupby(
    df['dateAdded'].dt.month
).agg({
    'id': 'count'
})

result
```

```python
df['dateUpdated'] = pd.to_datetime(df['dateUpdated'])
df['update_delta'] = (df['dateUpdated'] - df['dateAdded']).apply(lambda x: x.days)
result = df.groupby(
    df['city'], as_index=False
).agg({
    'update_delta': 'mean',
    'latitude': 'max'
})
zep_mean = float(result.loc[result['city'] == "Zephyrhills", "update_delta"])
zep_mean
```
